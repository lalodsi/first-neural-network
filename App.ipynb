```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
```

# Crear el dataset

Se utilizará la función `make_cicles` para generar una nube de puntos aleatorios con forma de círculo, esta nube se encontrará ligeramente dispersa y a simple vista se verán ambos círculos

```python
# Crear el dataset

n = 500
p = 2

[X, Y] = make_circles(n_samples=n, factor=0.5, noise=0.05)

plt.scatter(X[Y == 0,0], X[Y == 0,1], c='#aa336d')
plt.scatter(X[Y == 1,0], X[Y == 1,1], c='#70beb0')
plt.axis("equal")
plt.show()
```

## Capa de red neuronal

Se definirá una clase que se encargue de generar las capas. Así se podrá llamar a esta clase cuantas veces se deseen

```python
class neural_layer():
    
    def __init__(self, n_conn, n_neur, act_f):
        self.act_f = act_f
        self.b = np.random.rand(1, n_neur) * 2 - 1
        self.w = np.random.rand(n_conn, n_neur) * 2 - 1
```

```python
# Creando capas neuronales

topology = [p, 4, 8, 16, 8, 4, 1]

def create_nn(topology, act_f):
    nn = []
    for l, layer in enumerate(topology[:-1]):
        nn.append(neural_layer(topology[l], topology[l+1], act_f))
        
    return nn
```
