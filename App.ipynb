```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
```

# Crear el dataset

Se utilizará la función `make_cicles` para generar una nube de puntos aleatorios con forma de círculo, esta nube se encontrará ligeramente dispersa y a simple vista se verán ambos círculos

```python
# Crear el dataset

n = 500
p = 2

[X, Y] = make_circles(n_samples=n, factor=0.5, noise=0.05)

plt.scatter(X[Y == 0,0], X[Y == 0,1], c='#aa336d')
plt.scatter(X[Y == 1,0], X[Y == 1,1], c='#70beb0')
plt.axis("equal")
plt.show()
```
